```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
budget_data = {
    'Department': ['HR', 'IT', 'Finance', 'Marketing', 'Operations'],
    'Budget': [50000, 120000, 90000, 75000, 110000]
}
df_budget = pd.DataFrame(budget_data)
```

```python
expenses_data = {
    'Department': ['HR', 'IT', 'Finance', 'Marketing', 'Operations'],
    'Actual_Expenses': [52000, 115000, 95000, 80000, 108000]
}
df_expenses = pd.DataFrame(expenses_data)
```

```python
df = pd.merge(df_budget, df_expenses, on='Department')
```

```python
df['Variance'] = df['Actual_Expenses'] - df['Budget']
```

```python
df['Exceeding_Budget'] = df['Variance'] > 0
```

```python
df = df.sort_values(by='Variance', ascending=False)
```

```python
print(df)
```

```python
df.plot(x='Department', y=['Budget', 'Actual_Expenses'], kind='bar', figsize=(8,5))
plt.title('Department Budget vs Actual Expenses')
plt.ylabel('Amount')
plt.show()
```
