```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/student_marks.csv")

plt.figure(figsize=(8,5))
plt.bar(df['Subject'], df['Average_Marks'], color=['skyblue', 'lightgreen', 'salmon', 'orange', 'violet'])

plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.title("Comparison of Average Marks in Different Subjects")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/sales_category.csv")

plt.figure(figsize=(8,5))
plt.bar(df['Category'], df['Sales'], color=['skyblue', 'lightgreen', 'salmon', 'gold', 'violet'])

plt.xlabel("Product Category")
plt.ylabel("Sales")
plt.title("Sales Distribution Across Product Categories")

plt.show()
```
